```python
import pandas as pd
import xlsxwriter
import numpy as np
```

# Extracción de datos

Sacamos los datos de la hoja "Catálogo Truper". Dicha hoja debe ser la segunda dentro del excel

```python
df1 = pd.read_excel("Modelo Financiero.xlsx", sheet_name = 1)
df1.head()
```

Sacamos los datos del renglonaje, de la hoja "Catalogo Truper" (ahí se encuentra la columna "Total que paga Eddy a Truper con IVA). La hoja debe ser la primera dentro del excel

```python
df2 = pd.read_excel("Renglonaje.xlsx", sheet_name = 1)
df2.head()
```

# Limpieza de datos

```python
df1.replace("*", 0, inplace = True)
df2.replace("*", 0, inplace = True)
```

```python
df2.rename(columns={'Total que paga Edy a Truper contando reserva en garantía CON IVA':'Total Truper con IVA'}, inplace=True)
df1.rename(columns={'clave':'Clave'}, inplace=True)
```

# Creación del DF principal

Extreamos el costo total que Eddy paga a truper y lo anexamos al catálogo de Truper

```python
costo_t = df2[['Clave','Total Truper con IVA']]
df = df1.merge(costo_t, how='left', on='Clave')
df.head()
```

Limpiamos los datos de la columna anterior

```python
df["Total Truper con IVA"].replace("No disponible", 0, inplace = True)
df["Total Truper con IVA"] = df["Total Truper con IVA"].astype(float)
```

# Cálculos de la tabla final

```python
df["Precio Mayoreo con IVA Junio 2025"] = df['precio mayoreo con IVA']
df["Precio en linea con IVA"] = df['Precio Mayoreo con IVA Junio 2025']
df["Precio en linea sin IVA"] = df['Precio en linea con IVA'] / 1.16
```

*Nota: Dado que no sabemos el costo de envío, se quedará a 60 como placeholder

```python
df['Envio que se le cobra al cliente'] = np.nan
df['Envio que se le cobra al cliente'].fillna(60, inplace = True)
```

Más cálculos

```python
df['IVA en absolutos del producto que FERREPRO retiene en la compra'] = df['precio mayoreo con IVA'] - df['Precio en linea sin IVA']
df['Envio sin IVA para el consumidor que FERREPRO retiene'] = df['Envio que se le cobra al cliente'] / 1.16
df['IVA en absolutos del envio que se retiene'] = df['Envio que se le cobra al cliente'] - df['Envio sin IVA para el consumidor que FERREPRO retiene']
df['Total de ingreso con IVA (producto + envio)'] = df['precio mayoreo con IVA'] + df['Envio que se le cobra al cliente']
df['Total de ingreso sin IVA (producto + envio)'] = df['Total de ingreso con IVA (producto + envio)'] - df['IVA en absolutos del envio que se retiene'] - df['IVA en absolutos del producto que FERREPRO retiene en la compra']
```

*Nota: Dado que no sabemos el costo total de envío, se quedará en 80 como placeholder
- (¿Datos de Skydropx?)

```python
df['Total de envio con IVA'] = np.nan
df['Total de envio con IVA'].fillna(80, inplace = True)
```

Más cálculos

```python
df["Envio comprado sin IVA"] = df['Total de envio con IVA'] / 1.16
df["IVA en absolutos del envio"] = df['Total de envio con IVA'] - df['Envio comprado sin IVA']
```

*Nota: Dado que desconozco el criterio de cómo y cúando se aplican las comisiones, se quedarán como placeholders los valores de 0.16, 0 y 0 respectivamente

```python
df['Comision 1'] = np.nan
df['Comision 1'].fillna(0.16, inplace = True)

df['Comision 2'] = np.nan
df['Comision 2'].fillna(0, inplace = True)

df['Comision 3'] = np.nan
df['Comision 3'].fillna(0, inplace = True)

comisiones = df['Comision 1'] + df['Comision 2'] + df['Comision 3']
```

Más cálculos

```python
df['Tarifa por referencia que cobra Marketplace con IVA'] = df['Total de ingreso con IVA (producto + envio)'] * comisiones
df['Tarifa de referencia de Marketplace sin IVA'] = df['Tarifa por referencia que cobra Marketplace con IVA'] / 1.16
df['Iva en absolutos de la tarifa de referencia'] = df['Tarifa por referencia que cobra Marketplace con IVA'] - df['Tarifa de referencia de Marketplace sin IVA']
```

```python
df['Costo con IVA a pagar a TRUPER'] = df['Total Truper con IVA']
df['Costo sin IVA a pagar a TRUPER'] = df['Costo con IVA a pagar a TRUPER'] / 1.16
df['IVA del costo del producto pagado a TRUPER'] = df['Costo con IVA a pagar a TRUPER'] - df['Costo sin IVA a pagar a TRUPER']
```

```python
costoDelProducto = df['Costo con IVA a pagar a TRUPER']
envio = df['Total de envio con IVA']
TarifaDeAmazon = df['Tarifa por referencia que cobra Marketplace con IVA']
df['Total de egresos con IVA (costo de producto + envio + tarifa de amazon)'] = costoDelProducto + envio + TarifaDeAmazon
```

```python
envioSinIva = df['Envio comprado sin IVA']
tarifaSinIva = df['Tarifa de referencia de Marketplace sin IVA']
ivaDelProducto = df['IVA del costo del producto pagado a TRUPER']
df['Total de egresos sin IVA'] = envioSinIva + tarifaSinIva + ivaDelProducto
```

```python
df['Bruto'] = df['Total de ingreso con IVA (producto + envio)'] - df['Total de egresos con IVA (costo de producto + envio + tarifa de amazon)']
```

# Creación del archivo .xlsx

```python
writer = pd.ExcelWriter('Modelo Financiero py.xlsx', engine='xlsxwriter')
df.to_excel(writer, sheet_name='Modelo de rentabilidad', index=False)
```

```python
workbook = writer.book
worksheet = writer.sheets['Modelo de rentabilidad']
currency_format = workbook.add_format({'num_format': '$#,##0.00'})
text_format = workbook.add_format()
text_format.set_font('Arial')
worksheet.set_column('A:AY', 15, text_format)
worksheet.set_column('J:J', 15, currency_format)
worksheet.set_column('L:Q', 15, currency_format)
worksheet.set_column('S:T', 15, currency_format)
worksheet.set_column('AA:AM', 15, currency_format)
worksheet.set_column('AQ:AY', 15, currency_format)
```

```python
color1 = workbook.add_format({'bg_color': 'red'})
color2 = workbook.add_format({'bg_color': 'green'})
worksheet.conditional_format('AY2:AY15753', {'type':'cell', 'criteria': '<', 'value':0, 'format': color1})
worksheet.conditional_format('AY2:AY15753', {'type':'cell', 'criteria': '>=', 'value':0, 'format': color2})
```

```python
writer.close()
```
